```python
#############################################################################
#
#  csc531 Week5 -- Tutorial on Principal Component Analysis 
#  Week5_PCA.py ... written by Chakarida Nukoolkit (adapted from Dr.Jason Brownlee)
#            -- The sample code below do PCA to transform 4x5 --> 4x4  or 
#            from 5 columns, down to 4 columns...
#
#  Task 5-1 -- Step1 Run the given code below, then observe the result, 
#            especially the parameters of PCA and decomposition methods
#
#  Task 5-2 -- Make change to the array A to have more dimensions and 
#            more data, then try to do PCA to reduce it to 3, 2, and 1 component
#            Note !!! -- observe the resulted PCA and transformed data 
#            values and dimensions
#
##############################################################################
#
#  Principal Component Analysis
from numpy import array
from sklearn.decomposition import PCA
# define a matrix
A = array([[1, 2, 3, 2, 5], [3, 4, 5, 7, 6], [6, 7, 8, 9, 14], [0, 1, 2, 3, 4]])
print()
print("A ... original")
print(A)
# create the PCA instance of 4 components
pca = PCA(4)
# performing PCA by fitting on data
pca.fit(A)

# access PCA values and vectors
print()
print("pca.components:")
print(pca.components_)
print("\npca.explained_variance:")
print(pca.explained_variance_)
print("\npca.explained_variance_ratio:")
print(pca.explained_variance_ratio_)

# transform data and print new data
B = pca.transform(A)
print("\nA after got transformed:")
print(B)
```
