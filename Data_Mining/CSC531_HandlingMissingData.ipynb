```python
#-----------------------------------
# csc531 Week3 -- Part1 --Try to read data using url & Print out 
#                         descriptive statistics using "pandas"
# Ref:1. https://machinelearningmastery.com/load-machine-learning-data-python/
#     2. https://www.geeksforgeeks.org/working-with-missing-data-in-pandas/
#     3. https://www.kaggle.com/rtatman/data-cleaning-challenge-handling-missing-values
#     4. https://www.listendata.com/2019/06/pandas-read-csv.html (Read file CSV)
#     5. https://www.bogotobogo.com/python/scikit-learn/scikit_machine_learning_Data_Preprocessing-Missing-Data-Categorical-Data.php
#     6. https://www.pluralsight.com/guides/interpreting-data-using-descriptive-statistics-python (Mean, Mode, etc. of column)
# ---------------------------------
import pandas as pd

url2 = "https://raw.githubusercontent.com/jodyabney/R-Data-Analysis-Cookbook_Downloads/master/Chapter1/data/missing-data.csv"
data = pd.read_csv(url2, header =0)

print("\nMobile phone dataset --> Data dimension :")
print(data.shape)

print("Show ALL data :")
print(data)

print("\nOverall Basic-statistics :")
print(data.describe())
print("\nData info: ")
print(data.info())
print("\nMean or AVG of all columns :")
print(data.mean())
print("\nMode of all columns :")
print(data.mode())

#to get a mean or a mode of a particular column
print("\nMean of 'Income' : ")
print(data.loc[:,'Income'].mean())
print("\nMode of 'Phone_type': ")
print(data.loc[:,'Phone_type'].mode())
#-----------------------------------
# -- Part2 --Try to check and count for missing data or "NaN" with isnull()
#   
# ---------------------------------           
# displaying data only with plas = NaN 
print("\nCount the number of missing data points in each attr:")
missing_values_count = data.isnull().sum()
print(missing_values_count[0:3])
#
# ---------------------------------     
#    Task1 --> a) replace NaN of 'Income' with the column AVG
#              b) replace Nan of 'Phone_type' with the column Mode
# 
#     Your codes go here !!!
#
# ---------------------------------     
# Check for NaN after replacing missing data with specific values
print("\nAfter replacing NaN values --> Count the missing data points: ")
missing_values_count = data.isnull().sum()
print(missing_values_count[0:3])
print("\nAfter replacing NaN values --> Show all data: ")
print(data)
```
